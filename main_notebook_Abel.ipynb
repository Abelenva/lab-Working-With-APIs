# Use this notebook for your 3 choices of APIs

```python
import requests
import pandas as pd
```

```python
#coindesk
#It is useful for a project like creating a dashboard for my self to track prices and the market.
```

```python
url = "https://api.coindesk.com/v1/bpi/currentprice.json"
response = requests.get(url)
data = response.json()
bpi_data = data['bpi']
df = pd.DataFrame.from_dict(bpi_data, orient='index')

df
```

```python
#World time API
#In the same project I could keep a very close track of the time in any part of the world in order to have a complete dashboard to analyze the market
timezone = 'Europe/Berlin'  
url = f'http://worldtimeapi.org/api/timezone/{timezone}'
response = requests.get(url)
data = response.json()
time_df = pd.DataFrame([data])
time_df

```

```python
#realmealDB recipe API
#I could use this API to build and application that helps me cook with the ingredients I have left on the fridge.
```

```python
ingredient = 'chicken' 
url = f'https://www.themealdb.com/api/json/v1/1/filter.php?i={ingredient}'

response = requests.get(url)
data = response.json()

recipes = data['meals']
recipes_df = pd.DataFrame(recipes)

recipes_df[['strMeal', 'strMealThumb']]
```
